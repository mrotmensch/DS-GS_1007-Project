```python
import pandas as pd
import numpy as np
import urllib2
import csv
from pygeocoder import Geocoder
from pygeolib import GeocoderError
import math
```

the following functions read and clean the data in dataframe

getData: 
supposed to read the data from url into a csv

ConvertCSV_DF: 
converts csv file to DataFrame

cleanData:
1. only takes columns used
2. converts boro into full names
3. drops any rows without lat or long data
4. changes all other missing values to Unknown

```python
def getData(url):
    response = urllib2.urlopen(url)
    f = csv.reader(response)
    
    return f

def ConvertCSV_DF(f):
    df = pd.read_csv(f)
    
    return df

def cleanData(df):
    columns = ['Boro','Type','Provider','Name','Location','Lat','Long_','X','Y','Location_T','City','SSID']
    df_cut = df[columns]
    boro = {'BK': 'Brooklyn', 'MN': 'Manhattan', 'SI':'Staten Island', 'QU':'Queens', 'BX':'Bronx'}
    df_cut['Borough'] = df_cut['Boro'].apply(lambda name: boro[name])
    df_cut.drop('Boro',1, inplace = True)
    df_dropNaN = df_cut.dropna(subset = ['Lat','Long_'])
    df_new = df_dropNaN.replace(np.nan,"Unknown")
    
    return df_new
```

```python
df = ConvertCSV_DF('NYC_Free_Public_Wifi.csv')
df = cleanData(df)
df
```

the following functions convert a street address to lat and long coordinates

user_location:
set state default to New York (for now)
address is the street address entered by a user (there can be minor mistakes in this, pygeocoder takes into account this)

get_coordinates:

```python
def user_location(address):
    state = "New York"
    address_state = Geocoder.geocode(address + state)
    if address_state.valid_address == False:
        raise GeocodeError
    else:
        return address_state.formatted_address
    
def get_coordinates(user_address):
    address = Geocoder.geocode(user_address)
    return address.coordinates
```

the following function calculates straight distance between two coordinates (taken from the web)
calculates in miles

```python
def distance(lat1, long1, lat2, long2):

    # Convert latitude and longitude to 
    # spherical coordinates in radians.
    degrees_to_radians = math.pi/180.0
        
    # phi = 90 - latitude
    phi1 = (90.0 - lat1)*degrees_to_radians
    phi2 = (90.0 - lat2)*degrees_to_radians
        
    # theta = longitude
    theta1 = long1*degrees_to_radians
    theta2 = long2*degrees_to_radians
        
    # Compute spherical distance from spherical coordinates.
        
    # For two locations in spherical coordinates 
    # (1, theta, phi) and (1, theta, phi)
    # cosine( arc length ) = 
    #    sin phi sin phi' cos(theta-theta') + cos phi cos phi'
    # distance = rho * arc length
    
    cos = (math.sin(phi1)*math.sin(phi2)*math.cos(theta1 - theta2) + 
           math.cos(phi1)*math.cos(phi2))
    arc = math.acos( cos ) * 3960

    # Remember to multiply arc by the radius of the earth 
    # in your favorite set of units to get length.
    return arc
```
